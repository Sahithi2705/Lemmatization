Downloading English Model

```python
import stanza
stanza.download('en')
```

Loading Stanza Pipeline

```python
nlp = stanza.Pipeline('en', processors='tokenize,pos,lemma')
```

Pragraph

```python
paragraph = """
The students were running in the playground and buying snacks after school.
They were playing games and enjoying their evening together.
Some students were studying while others were talking happily.
Teachers were observing the students and guiding them properly.
Everyone felt excited and motivated during the activities.
"""

print("Original Paragraph:\n")
print(paragraph)
```

Lemmatization using Stanza

```python
doc = nlp(paragraph)

lemmas = []

for sentence in doc.sentences:
    for word in sentence.words:
        lemmas.append(word.lemma.lower())

lemmatized_text = " ".join(lemmas)

print("\nLemmatized Text:\n")
print(lemmatized_text)
```

Count Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
X = vectorizer.fit_transform([lemmatized_text])

vocab = vectorizer.get_feature_names_out()
count_matrix = X.toarray()

print("\nVocabulary:\n", vocab)
print("\nCount Vector:\n", count_matrix)
```

Term Frequency

```python
import pandas as pd

total_words = count_matrix.sum()

tf_values = [count/total_words for count in count_matrix[0]]

tf_df = pd.DataFrame({
    "Word": vocab,
    "Count": count_matrix[0],
    "TF": tf_values
})

print("\nTerm Frequency Table:\n")
print(tf_df)
```
